```
import networkx as nx  #not optimal
#'r' 'g' 'b' 'y' 'm' 'c' 'k'
G = nx.Graph()
G.add_edge(0,4,color='r',weight=6)#0
G.add_edge(0,1,color='g',weight=6)#1
G.add_edge(0,2,color='b',weight=6)#2
G.add_edge(1,5,color='r',weight=6)#3
G.add_edge(1,2,color='c',weight=6)#4
G.add_edge(2,5,color='k',weight=6)#5
G.add_edge(3,0,color='y',weight=6)#6
G.add_edge(3,2,color='r',weight=6)#7
G.add_edge(3,1,color='m',weight=6)#8
G.add_edge(3,4,color='g',weight=6)#9
G.add_edge(4,1,color='b',weight=6)#10
G.add_edge(4,2,color='y',weight=6)#11
G.add_edge(5,4,color='m',weight=6)#12

colors = nx.get_edge_attributes(G,'color').values()
weights = nx.get_edge_attributes(G,'weight').values()

pos = nx.circular_layout(G)
nx.draw(G, pos, 
        edge_color=colors, 
        width=list(weights),
        with_labels=True,
        node_color='lightgreen')
```
